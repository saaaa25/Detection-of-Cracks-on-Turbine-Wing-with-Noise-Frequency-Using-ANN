```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
data = pd.read_csv("dataset.csv")
data = data.drop(['Index'], axis =1)
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,7:]
Y = data.iloc[:,0:7]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
Y1_train = Y_train.iloc[:,0]
Y2_train = Y_train.iloc[:,1]
Y3_train = Y_train.iloc[:,2]
Y4_train = Y_train.iloc[:,3]
Y5_train = Y_train.iloc[:,4]
Y6_train = Y_train.iloc[:,5]
Y7_train = Y_train.iloc[:,6]

Y1_test = Y_test.iloc[:,0]
Y2_test = Y_test.iloc[:,1]
Y3_test = Y_test.iloc[:,2]
Y4_test = Y_test.iloc[:,3]
Y5_test = Y_test.iloc[:,4]
Y6_test = Y_test.iloc[:,5]
Y7_test = Y_test.iloc[:,6]
```

```python
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
def scale_datasets(x_train, x_test):
  """
  Standard Scale test and train data
  Z - Score normalization
  """
  standard_scaler = StandardScaler()
  x_train_scaled = pd.DataFrame(
      standard_scaler.fit_transform(x_train),
      columns=x_train.columns
  )
  x_test_scaled = pd.DataFrame(
      standard_scaler.transform(x_test),
      columns = x_test.columns
  )
  return x_train_scaled, x_test_scaled
x_train_scaled, x_test_scaled = scale_datasets(X_train, X_test)
```

# Lasso Regression

The LASSO (Least Absolute Shrinkage and Selection Operator) model is a linear model that uses regularisation and variable selection to produce a more precise and interpretable statistical model. The acronym LASSO sometimes goes by the name "Least Absolute Shrinkage and Selection Operator." The L1 penalty and the mean squared error loss function are both optimized in this model. The L1 penalty is proportional to the total number of model parameters. Having a regularising effect, causes the model's parameters to converge toward zero, making the model sparser.

```python
# train the model
model1 = linear_model.Lasso(alpha = 0.006725, precompute = True)
model1.fit(x_train_scaled, Y1_train)
ypred1 = model1.predict(x_test_scaled)
X_test['L1_prediction'] = ypred1

print("Mean Squared Error : ", mean_squared_error(Y1_test, ypred1))
print("Mean Absolute Error : ", mean_absolute_error(Y1_test, ypred1))
```
